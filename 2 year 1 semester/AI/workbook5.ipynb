Задание:
Создайте класс по работе с тригонометрическими функциями. В классе должны быть реализованы функции вычисления:
+ косинуса;
+ синуса;
+ тангенса;
+ арксинуса;
+ арккосинуса;
+ арктангенса;
+ перевода из градусов в радианы.

```python
import math

class Trigonometry:
    def __init__(self):
        pass

    @staticmethod
    def to_radians(degrees):
        return degrees * (math.pi / 180)

    def cos(self, angle_degrees):
        radians = self.to_radians(angle_degrees)
        return math.cos(radians)

    def sin(self, angle_degrees):
        radians = self.to_radians(angle_degrees)
        return math.sin(radians)

    def tan(self, angle_degrees):
        radians = self.to_radians(angle_degrees)
        return math.tan(radians)

    def arcsin(self, value):
        if -1 <= value <= 1:
            radians = math.asin(value)
            return radians * (180 / math.pi)
        else:
            raise ValueError("Значение должно быть в диапазоне [-1, 1]")

    def arccos(self, value):
        if -1 <= value <= 1:
            radians = math.acos(value)
            return radians * (180 / math.pi)
        else:
            raise ValueError("Значение должно быть в диапазоне [-1, 1]")

    def arctan(self, value):
        radians = math.atan(value)
        return radians * (180 / math.pi)

if __name__ == "__main__":
    calc = Trigonometry()

    print("Перевод градусов в радианы:")
    print(f"45° = {calc.to_radians(45):.4f} рад")
    print(f"90° = {calc.to_radians(90):.4f} рад")
    print(f"180° = {calc.to_radians(180):.4f} рад")

    print("\nТригонометрические функции:")
    print(f"cos(60°) = {calc.cos(60):.4f}")
    print(f"sin(30°) = {calc.sin(30):.4f}")
    print(f"tan(45°) = {calc.tan(45):.4f}")
    print(f"arcsin(0.5) = {calc.arcsin(0.5):.4f}°")
    print(f"arccos(0.5) = {calc.arccos(0.5):.4f}°")
    print(f"arctan(1.0) = {calc.arctan(1.0):.4f}°")
```

Дан класс, описывающий бинарное дерево.

Реализуйте в классе функцию для вставки нового элемента в дерево по следующим правилам:

•	Левое поддерево узла содержит только узлы со значениями меньше, чем значение в узле.

•	Правое поддерево узла содержит только узлы со значениями меньше, чем значение в узле.

•	Каждое из левого и правого поддеревьев также должно быть бинарным деревом поиска.

•	Не должно быть повторяющихся узлов.

Метод вставки сравнивает значение узла с родительским узлом и решает куда добавить элемент (в левое или правое поддерево). Перепишите, метод PrintTree для печати полной версии дерева.

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Tree(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Tree(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def _print_tree(self, level=0, is_right=None):
        """
        Рекурсивный метод для печати дерева в виде горизонтальной структуры.
        """
        if self.right:
            self.right._print_tree(level + 1, True)

        prefix = " " * 4 * level  # Индентация для уровня дерева
        if is_right is None:  # Корневой узел
            print(f"{prefix}{self.data}")
        else:
            direction = "└── " if is_right else "┌── "
            print(f"{prefix}{direction}{self.data}")

        if self.left:
            self.left._print_tree(level + 1, False)

    def print_tree(self):
        self._print_tree()


# Создаем дерево и вставляем элементы
root = Tree(12)
root.insert(1)
root.insert(14)
root.insert(2)
root.insert(3)
root.insert(4)
root.insert(11)
root.insert(5)
root.insert(6)
root.insert(15)
root.insert(13)

# Печатаем дерево
root.print_tree()
```

### Задание
Постройте классификатор на основе дерева принятия решений следующего датасета:

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Исходные данные
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = np.array([0, 0, 0, 1, 1, 1])

# Создаем и обучаем классификатор
classifier = DecisionTreeClassifier()
classifier.fit(X, target)

# Предсказываем значения
y_pred = classifier.predict(X)

# Выводим Confusion Matrix и Classification Report
print("Confusion Matrix:")
print(confusion_matrix(target, y_pred))
print("\nClassification Report:")
print(classification_report(target, y_pred))

# Визуализируем дерево решений
plt.figure(figsize=(15, 10))
plot_tree(classifier, filled=True, feature_names=['X1', 'X2'], class_names=['0', '1', '2']) # Добавляем имя третьего класса
plt.show()
```
