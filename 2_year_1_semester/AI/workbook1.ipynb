```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем массив случайных значений из интервала (0; 1)
array_size = 100  # Размер массива
random_array = np.random.rand(array_size)

# Рассчитываем среднее значение
mean = np.mean(random_array)

# Рассчитываем медианное значение
median = np.median(random_array)

print(f"Среднее значение: {mean:.4f}")
print(f"Медианное значение: {median:.4f}")

# Строим точечную диаграмму рассеяния
plt.figure(figsize=(8, 6))
plt.scatter(range(array_size), random_array, s=50, c='blue', alpha=0.7)
plt.title("Точечная диаграмма рассеяния случайных значений")
plt.xlabel("Индекс")
plt.ylabel("Значение")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x)**2)) / (np.abs(1 - np.sin(x)**3))

# Создаем массив x
x = np.linspace(1, 10, 10)

# Вычисляем значения функции
y = f(x)

# Выделяем срез
y_slice = y[:5]
x_slice = x[:5]

# Строим графики
plt.figure(figsize=(10, 6))

# Линейный график для основного массива
plt.plot(x, y, marker='o', linestyle='-', label='f(x)')

# Точечный график для среза (на том же графике)
plt.scatter(x_slice, y_slice, s=200, c='purple', label='Срез')

plt.title('График функции и срез')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid

def f(x):
  return abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 11, 1)
y = f(x)

# Вычислить площадь под кривой с помощью правила трапеций
area = trapezoid(y, x)

# Создать график
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='purple')
plt.fill_between(x, y, color='lightgrey')
plt.title('График функции и площадь под ней')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.text(0.1, 0.1, f'Площадь: {area:.4f}', fontsize=14)

plt.show()
```

```python
import math
def calculator():
    """
    Простейший калькулятор, выполняющий базовые арифметические операции
    и вычисляющий функции для двух переменных.
    """

    while True:
        print("Выберите операцию:")
        print("1. Сложение (+)")
        print("2. Вычитание (-)")
        print("3. Умножение (*)")
        print("4. Деление (/)")
        print("5. e^(x+y)")
        print("6. sin(x+y)")
        print("7. cos(x+y)")
        print("8. x^y")
        print("9. Выход")

        choice = input("Введите номер операции (1/2/3/4/5/6/7/8/9): ")

        if choice in ('1', '2', '3', '4', '5', '6', '7', '8'):
            try:
                x = float(input("Введите первое число (x): "))
                y = float(input("Введите второе число (y): "))
            except ValueError:
                print("Ошибка: Введите корректные числа.")
                continue

            if choice == '1':
                print(x, "+", y, "=", x + y)
            elif choice == '2':
                print(x, "-", y, "=", x - y)
            elif choice == '3':
                print(x, "*", y, "=", x * y)
            elif choice == '4':
                if y == 0:
                    print("Ошибка: Деление на ноль!")
                else:
                    print(x, "/", y, "=", x / y)
            elif choice == '5':
                print("e^(", x, "+", y, ") =", math.exp(x + y))
            elif choice == '6':
                print("sin(", x, "+", y, ") =", math.sin(x + y))
            elif choice == '7':
                print("cos(", x, "+", y, ") =", math.cos(x + y))
            elif choice == '8':
                print(x, "^", y, "=", x**y)

        elif choice == '9':
            break
        else:
            print("Ошибка: Неверный выбор операции.")

if __name__ == "__main__":
    calculator()
```
